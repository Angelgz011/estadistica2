# Introducción
En este trabajo tiene como objetivo buscar alguna relación entre las condiciones admosféricas y sus componentes como causantes de la contaminación de las ciudades. Se busca comparar estas variables con la Cálidad del Aire (AQI) y determinar cuales variables meteorológicas tienen mayor relación.
# Hipotesis

H0 (Hipótesis nula): Las variables meteorológicas (humedad, presión atmosférica, precipitación, visibilidad y latitud) no tienen un efecto significativo sobre el AQI.

H1 (Hipótesis alternativa): Al menos una de las variables meteorológicas tiene un efecto significativo sobre el AQI.
# Marco Teórico
Para el estudio de esta base de datos se hará uso de herramientas como ANOVA y Tukey, Histogramas, Mapa de correlación y el analisis de variables como el coeficiente de determinación y correlación.

# Metodología
Para el estudio de esta base de datos se hará uso de herramientas como ANOVA y Tukey, Histogramas, Mapa de correlación y el analisis de variables como el coeficiente de determinación y correlación.
Se filtrará primero los valores nulos, y se eliminan aquiellas columnas irrelevantes.
Se hace un analisis por medio de histogramas para visualizar la distribución de contaminantes.
Se aplican las pruebas a cada variables y se determina su relevancia para el estudio o el descarte o renombre de esta.
Regresión lineal múltiple:

Variable dependiente: AQI

Variables independientes: Humedad, presión, precipitación, visibilidad y latitud.

División del dataset en conjunto de entrenamiento (80%) y prueba (20%).

Evaluación del modelo mediante el coeficiente de determinación (R²).
# Resultados
La matriz de correlación muestra que no existen correlaciones fuertes entre el Índice de Calidad del Aire (AQI) y las variables meteorológicas consideradas (humedad, presión, precipitación, visibilidad y latitud). La mayor parte de los coeficientes de correlación se encuentran cercanos a cero, lo que sugiere una relación débil o nula entre dichas variables y el AQI.

Regresión Lineal Múltiple
El modelo de regresión lineal múltiple arrojó un coeficiente de determinación (R²) de -0.0110, lo cual indica que el modelo no solo no logra explicar la variabilidad del AQI a partir de las variables seleccionadas, sino que su ajuste es peor que el de un modelo constante (promedio). Asimismo, el coeficiente de correlación es de 0.0001, confirmando la ausencia de una relación lineal significativa entre las variables predictoras y la variable dependiente.

Análisis de Varianza (ANOVA)
Los valores de p (PR(>F)) para todas las variables independientes superan ampliamente el umbral de significancia estadística común (0.05):

Humedad: p = 0.4096

Presión: p = 0.9341

Precipitación: p = 0.2312

Visibilidad: p = 0.8707

Latitud: p = 0.6309

Estos resultados indican que ninguna de las variables meteorológicas incluidas tiene un efecto significativo sobre el AQI en este modelo.
# Conclusión
En conclsuión los análisis realizados indican que las variables meteorológicas seleccionadas no tienen un efecto significativo sobre la Calidad del Aire en el conjunto de datos analizado.
Esto sugiere que si bien, las condiciones meteorológicas pueden influir en la dispersión de contaminantes atmosféricos en ciertas cirscunstancias, no son predictoras sólidas del AQI por si solas.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
nivel_de_significancia = 0.05
tukey = pairwise_tukeyhsd(endog = df['PM25'], groups = df['Season'], alpha = nivel_de_significancia)
tukey.plot_simultaneous(ylabel = 'Niveles de medición', xlabel = 'Unidades de␣la variable')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.title('')
print(tukey)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm

df = df.rename(columns={"Humidity (%)": "Humidity"})
df = df.rename(columns={"Pressure (hPa)": "Pressure"})
df = df.rename(columns={"Visibility (km)": "Visibility"})
df = df.rename(columns={"Precipitation (mm)": "Precipitation"})

X =df[['Humidity','Pressure','Precipitation','Visibility','Latitude']]
Y = df['AQI']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state = 42)
X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)
modelo = sm.OLS(Y_train, X_train_constante).fit()

Yc = modelo.predict(X_test_constante)
plt.scatter(Y_test,Yc)
r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinación: {r2: 0.4f}')
coeficiente_de_correlación = r2**2
print(f'Coeficiente de correlación: {coeficiente_de_correlación: 0.4f}')
```
